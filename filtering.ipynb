```python
import cv2
import numpy as np
from keras.datasets import cifar10
import matplotlib.pyplot as plt

input_range = 10
sigma = [1, 5, 10, 50, 100]

# To store images from cifar10 10 classes
input_image_dataset = []

# To load the CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# To get the first image from every class in the training set
for i in range(input_range):
  first_image_index = next((index for index, label in enumerate(y_train) if label[0] == i), None)
  input_image_dataset.append(x_train[first_image_index])

# To plot original images
figure = plt.figure(figsize = (32, 32))
figure.tight_layout()

for i in range(input_range):
  plt.subplot(8, 10, i + 1)
  plt.imshow(input_image_dataset[i])
  plt.title("Original Image " + str(i+1))

# Gaussian filter
for i in range(input_range):
  for j, sigma_val in enumerate(sigma):
    plt.subplot(8, 10, (10 * (j + 1)) + i + 1)
    gaussian_filtered_image = cv2.GaussianBlur(input_image_dataset[i], (5, 5), sigma_val)
    plt.imshow(gaussian_filtered_image)
    plt.title("Gaussian Sigma {}".format(sigma_val))

# Moving average filter
for i in range(input_range):
  plt.subplot(8, 10, (i + 1) + 60)
  moving_avg_filtered_image = cv2.blur(input_image_dataset[i], (5, 5))
  plt.imshow(moving_avg_filtered_image)
  plt.title("Moving Average")

# Median filter
for i in range(input_range):
  plt.subplot(8, 10, (i + 1) + 70)
  median_filtered_image = cv2.medianBlur(input_image_dataset[i], 5)
  plt.imshow(median_filtered_image)
  plt.title("Median")
```
