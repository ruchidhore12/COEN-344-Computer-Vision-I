```python
import cv2 as cv
import numpy as np
from keras.datasets import cifar10
import matplotlib.pyplot as plt

input_range = 10

# To store images from cifar10 10 classes
input_image_dataset = []

# To load the CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# To get the first image from every class in the training set
for i in range(input_range):
  first_image_index = next((index for index, label in enumerate(y_train) if label[0] == i), None)
  input_image_dataset.append(x_train[first_image_index])

# To plot original images
figure = plt.figure(figsize = (32, 32))
figure.tight_layout()

for i in range(input_range):
  plt.subplot(8, 10, i + 1)
  plt.imshow(input_image_dataset[i])
  plt.title("Image " + str(i + 1))

# SIFT detector
sift_detector = cv.xfeatures2d.SIFT_create()

index = 1
for image in input_image_dataset:
    # To convert the image to grayscale
    gray_image = cv.cvtColor(image, cv.COLOR_RGB2GRAY)
    
    # To detect keypoints and compute descriptor using SIFT
    keypoints, descriptor = sift_detector.detectAndCompute(gray_image, None)
    
    # To draw keypoints on the image
    img_with_keypoints = cv.drawKeypoints(image, keypoints, outImage = np.array([]), flags = cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    
    # To display the image with the detected keypoints
    plt.subplot(8, 10, index + 10)
    plt.imshow(img_with_keypoints)
    index = index + 1
```
