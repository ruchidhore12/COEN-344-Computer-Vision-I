```python
import cv2
import numpy as np
from keras.datasets import cifar10
import matplotlib.pyplot as plt

input_range = 10
harris_corner_partner = [0.04, 0.07]
image_indexes = [10, 20]

# To store images from cifar10 10 classes
input_image_dataset = []

# To load the CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# To get the first image from every class in the training set
for i in range(input_range):
  first_image_index = next((index for index, label in enumerate(y_train) if label[0] == i), None)
  input_image_dataset.append(x_train[first_image_index])

# To plot original images
figure = plt.figure(figsize = (32, 32))
figure.tight_layout()

for i in range(input_range):
  plt.subplot(8, 10, i + 1)
  plt.imshow(input_image_dataset[i])
  plt.title("Original Image " + str(i+1))

# To calculate harris corner map
def harris_corner_detector(image, harris_parameter = 0.04):
  gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
  gray = np.float32(gray)
  corner_detection_map = cv2.cornerHarris(gray, 2, 3, harris_parameter)
  return corner_detection_map

# To plot harris corner map
def plot(corner_detection_map, harris_parameter = 0.04, index = 10):
  plt.subplot(8, 10, (i + 1) + index)
  figure.tight_layout()
  plt.imshow(corner_detection_map, cmap = 'gray')
  plt.title("Harris Corner Parameter {}".format(str(harris_parameter)))

# To provoke harris corner with different parameter values
for i in range(input_range):
  corner_detection_map = harris_corner_detector(input_image_dataset[i], harris_corner_partner[0])
  plot(corner_detection_map, harris_corner_partner[0], image_indexes[0])
  corner_detection_map = harris_corner_detector(input_image_dataset[i], harris_corner_partner[1])
  plot(corner_detection_map, harris_corner_partner[1], image_indexes[1])





```
